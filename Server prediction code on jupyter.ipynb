<h1>First selection sample data</h1>

```python
#singular = alldata.drop_duplicates(subset=["authors"], keep="first")
mostknown = alldata.sort_values(["work_ratings_count"],ascending= False )[alldata.work_ratings_count>100000]
authors = mostknown["authors"].value_counts().index.tolist()
firstauth = []
for au in authors:
    firstauth.append(mostknown[mostknown.authors == au].sample(frac=1)[0:3])
singular = pd.concat(firstauth)
selbooks = singular.sort_values(["work_ratings_count"],ascending= False )["book_id"][0:40]
selbooks = selbooks.sample(frac=1).values.tolist()
selbooks
```

<h1>User based elaboration</h1>

```python
#suppose that these are the selected books: (book_id)
selectedbooks = [2767052, 3, 5907, 7260188]
selectedbooksid = []
for book in selectedbooks:
    selectedbooksid.append(alldata[alldata.book_id == book]["id"].values[0])
positives = userbook[userbook.rating >=4]
print(selectedbooksid)
```

```python
#now I need to compare the selected books with the most liked books by folks:
intersect = set([])
for i in range(len(selectedbooksid)):
    if(i == 0):
        intersect = set(positives[positives.book_id == selectedbooksid[0]]["user_id"])
    else:
        intersect = intersect.intersection( set(positives[positives.book_id == selectedbooksid[i]]["user_id"]) )
#these users has the same background 
intersect = list(intersect)
similusers = pd.DataFrame(data={"user_id":intersect})
```

```python
#rating most uniform to population selection
ratingpoint= []
for userid in intersect:
    ratingpoint.append(userresume[userresume.user_id == userid]["ratingpoint"].values[0])

similusers = similusers.assign(ratingpoints = ratingpoint)
#take only 5 people
bestusers = similusers.sort_values(["ratingpoints"],ascending= True )["user_id"].values[0:3]
bestusers
```

```python
#now i have most similar person to my user, so I need to select books from him
bestintersect = set([])
for user in bestusers:
    if(user == bestusers[0]):
        bestintersect = set(positives[positives.user_id == user]["book_id"].values)
    else:
        bestintersect = bestintersect.intersection( set(positives[positives.user_id == user]["book_id"].values) )
    
bestintersect = list(bestintersect)
#remove the selected books
for b in selectedbooksid:
    bestintersect.remove(b)
bestintersect
```

```python
#let's see what books taste good
returndata = []
for bo in bestintersect:
    returndata.append(alldata[alldata.id == bo])
returndata = pd.concat(returndata)

returnid = returndata["book_id"].values.tolist()
returntitle = returndata[ "title"].values.tolist()
returntitle = [w.replace(',', '') for w in returntitle]
returnprice = returndata[ "price"].values.tolist()
returnauthor = returndata[ "authors"].values.tolist()
returnauthor = [w.replace(',', '') for w in returnauthor]

returnstr = ""
for i in range(len(returnid)):
    if(i !=0):
        returnstr += ","
    returnstr += str(returnid[i])+"/"+returntitle[i]+"/"+str(returnprice[i])+"/"+returnauthor[i]

print(returnstr)
```

<h1>Author based elaboration</h1>

```python
#get all books red by selected users
allbooksreaded = []
for user in bestusers:
    allbooksreaded.append( positives[positives.user_id == user]) 
allbooksreaded = pd.concat(allbooksreaded)


#drop the selected books
for i in selectedbooksid:
    allbooksreaded = allbooksreaded[allbooksreaded.book_id != i]

#get descriptions of all books
bookdesc = []
for book in allbooksreaded["book_id"]:
    bookdesc.append( alldata[alldata.book_id == book]) 
bookdesc = pd.concat(bookdesc)
auth = bookdesc["authors"].value_counts()
auth = auth[auth >1].index.values.tolist()

#now i get the seleced authors:
#drop the selected books
selectedauthors = []
for i in selectedbooksid:
    selectedauthors.append(alldata[alldata.id ==i]["authors"].values[0])

auth = auth + selectedauthors

auth = pd.Series(auth).drop_duplicates(keep="first").values.tolist()

#now I can clean the readed books
allbooksreaded = allbooksreaded.drop_duplicates(subset=["book_id"], keep="first")#.values.tolist()

#now put authors in userbook dataset for better selection 
aulist = []
for bid in allbooksreaded["book_id"].values.tolist():
    aulist.append(alldata[alldata.id == bid]["authors"].values[0])
allbooksreaded = allbooksreaded.assign(authors = aulist)

#select only 3 best authors' books
elabau = []
for au in auth:
    elabau.append(sklearn.utils.shuffle(allbooksreaded[allbooksreaded.authors == au])[0:3])
elabau = pd.concat(elabau)["book_id"].values.tolist()

outbook = []
for e in elabau:
    outbook.append(alldata[alldata.id == e])
outbook = pd.concat(outbook)
outbook
    
```
