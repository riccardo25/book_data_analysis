```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from bs4 import BeautifulSoup
import requests
import cv2
from sklearn.cluster import KMeans
```

```python
dataBooks = pd.read_csv("dataset\\books.csv")
dataBooks = dataBooks
dataBooks.head()
```

```python
dataBooks
```

```python
tags = pd.read_csv("dataset\\book_tags.csv")
tags.head()
```

```python
bookid = dataBooks["book_id"]
complessivo = pd.DataFrame(columns= ["goodreads_book_id","tag_id", "count"])

for id_id in bookid:
    complessivo = complessivo.append(tags[tags.goodreads_book_id == id_id][0:10], ignore_index=True)
    
#remove complessive
moretaggedid = complessivo.drop_duplicates("tag_id")["tag_id"].values.tolist()
moretaggedid
```

```python
tags = pd.read_excel("dataset\\new\\tagpulitixls.xls")#pd.read_csv("dataset\\tags.csv")
tags.head()
```

```python
tagiddata = pd.DataFrame(columns= ["tag_id", "occourances", "sum"])
for tagid in moretaggedid:
    c = complessivo[complessivo.tag_id == tagid]["count"]
    occ = c.count()
    summ = c.sum()
    tagiddata = tagiddata.append({"tag_id": tagid, "occourances":occ,  "sum" : summ}, ignore_index=True)
```

```python
list_trans = []
for idtag in tagiddata["tag_id"]:
    namelist = tags.loc[tags["tag_id"] == idtag, ["tag_name"]].values.tolist()
    if(namelist):
        list_trans.append(namelist[0][0])
    else:
        list_trans.append(None)
        
tagiddata["translation"] = list_trans
```

```python
outdata = tagiddata.mask(lambda x : x["occourances"]< 100).dropna()
```

```python
outdata.to_csv("dataset\\new\\elaboratedtags.csv")
```
