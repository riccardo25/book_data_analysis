```python
import winsound as ws
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from string import digits
```

<h1>Tag cleaning</h1>

```python
dataTags = pd.read_csv("dataset\\tags.csv")
dataTags
```

```python
#removing no english tags
def isEnglish(s):
    try:
        s.encode(encoding='utf-8').decode('ascii')
    except UnicodeDecodeError:
        return False
    else:
        return True

tagtable = dataTags.mask(dataTags["tag_name"].map(lambda x: not isEnglish(x))).dropna()
tagtable
```

```python
newtags = tagtable["tag_name"].map(lambda x: x.replace("-", " "))
for i in range(5):
    newtags = newtags.map(lambda x: x.replace("  ", " "))

def firstlastspace(stringa):
    try:
        if(stringa[0] == " "):
            stringa = stringa[1:]
        if(stringa[-1] == " "):
            stringa = stringa[:-1]
        return stringa
    except:
        return ""

newtags = newtags.map(lambda x: firstlastspace(x))
tagtable["tag_name"] = newtags
tagtable.head()
```

```python
#remove all numbers
#tagtable = tagtable.mask( tagtable["tag_name"].map(lambda x: x.isdigit())).dropna()
dig = str.maketrans('', '', digits)
tagtable["tag_name"] = tagtable["tag_name"].map(lambda x: x.translate(dig))
tagtable = tagtable.mask( tagtable["tag_name"].map(lambda x: len(x)<4 )).dropna()
tagtable
```
