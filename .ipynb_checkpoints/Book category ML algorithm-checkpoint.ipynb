```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('fivethirtyeight')
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline
from bs4 import BeautifulSoup
import requests
import cv2
from sklearn.cluster import KMeans
from sklearn.linear_model import LogisticRegression
from sklearn import tree
from sklearn.neural_network import MLPClassifier
from sklearn.naive_bayes import BernoulliNB
from sklearn.ensemble import RandomForestClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
alldata = pd.read_csv("dataset\\finals\\complissivolibricongenere.csv").drop(axis=1, columns=["Unnamed: 0"])
alldata = alldata.dropna()
alldata = alldata.drop(axis=1, columns=["gendernames", "color_name", "title", "authors"])
#shuffling data
alldata = alldata.sample(frac=1)
alldata.head()
```

<h1>Categorization based on ML algorithm</h1>

```python
N = 4500
trigger = 200
#get only most important class
data = alldata[alldata.genderclass == 0]
for c in alldata["genderclass"].value_counts().index :
    if(c != 0 and len(alldata[alldata.genderclass == c]) > trigger ):
        data = data.append(alldata[alldata.genderclass == c])
        
print("Libri considerati per la classificazione: "+str(len(data)))
print("Libri scartati: "+str(len(alldata)-len(data)))
print("Classi totali: "+str(len(alldata["genderclass"].value_counts().index)))
print("Classi considerate per la classificazione: "+str(len(data["genderclass"].value_counts().index)))
#shuffle
data = data.sample(frac=1)
x_train = data[:N].drop(axis=1, columns=["genderclass", "book_id", "id", "genderid"])
y_train = data[:N]["genderclass"]
x_eval  = data[N:10000].drop(axis=1, columns=["genderclass", "book_id", "id", "genderid"])
y_eval  = data[N:10000]["genderclass"].values.tolist()
```

<h2>Naive Bayes</h2>

```python
#logistic regression for clustering
model = GaussianNB()
model.fit(x_train, y_train)
```

```python
ynew = model.predict(x_eval)
print("RATE: "+str((ynew - y_eval).tolist().count(0)/len(ynew)))
for i in range(len(x_eval)):
    ;
    #print("X="+str(i)+", Predicted=" + str(ynew[i]) + "\treal:" + str(y_eval[i]) )
```

<h2>Decision Tree</h2>

```python
#logistic regression for clustering
model = tree.DecisionTreeClassifier()
model.fit(x_train, y_train)
```

```python
ynew = model.predict(x_eval)
print("RATE: "+str((ynew - y_eval).tolist().count(0)/len(ynew)))
for i in range(len(x_eval)):
    print("X="+str(i)+", Predicted=" + str(ynew[i]) + "\treal:" + str(y_eval[i]) )
```

<h2>Random forest</h2>

```python
model = RandomForestClassifier(max_depth=len(alldata["genderclass"].value_counts()), random_state=0)
model.fit(x_train, y_train)
```

```python
ynew = model.predict(x_eval)
print("RATE: "+str((ynew - y_eval).tolist().count(0)/len(ynew)))
for i in range(len(x_eval)):
    print("X="+str(i)+", Predicted=" + str(ynew[i]) + "\treal:" + str(y_eval[i]) )
```
